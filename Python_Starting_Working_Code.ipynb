```python
# Pixel average
# TOCOMPLETE
def pixel_average(p):
  return int((p[0] + p[1] + p[2])/ 3)
```

```python
# Replace bright pixels by grey pixels
# TOCOMPLETE
my_image = Image.open("cat.jpg")
my_image.thumbnail((250,250))
my_image
```

```python
def filter_average(img):
  for x in range(img.width):
    for y in range(img.height):
      pixel = img.getpixel((x,y))
      pixel_avg = pixel_average(pixel)
      if pixel_avg > 120:
        img.putpixel((x,y), (pixel_avg,pixel_avg,pixel_avg))

filter_average(my_image)
my_image
```

```python
# Generate a border around an image
img = Image.open("cat.jpg")
img.thumbnail((250,250))

border_size = 10
width=img.width + 2 * border_size
height=img.height + 2*border_size
new_img = Image.new("RGB", (width, height),color="black")
new_img.paste(img, (border_size, border_size))

new_img
```
